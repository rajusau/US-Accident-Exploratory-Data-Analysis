````python
import matplotlib.pyplot as plt


df['Start_Time'] = pd.to_datetime(df['Start_Time'])

df['Month'] = df['Start_Time'].dt.month

df = df.dropna(subset=['Temperature', 'Month'])

plt.figure(figsize=(10, 6))
plt.hist2d(df['Temperature'], df['Month'], bins=(20, 12), cmap=plt.cm.Blues)
plt.colorbar(label='Number of Accidents')
plt.xlabel('Temperature')
plt.ylabel('Month')
plt.title('Number of Accidents by Temperature and Month')
plt.xticks(rotation=45)
plt.yticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
                          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.tight_layout()
plt.show()
````

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df['Start_Time'] = pd.to_datetime(df['Start_Time'])

df['Day_of_Week'] = df['Start_Time'].dt.dayofweek

temperature_ranges = [
    (-50, 0),    # Very Cold
    (0, 32),     # Freezing
    (32, 50),    # Cold
    (50, 68),    # Cool
    (68, 80),    # Warm
    (80, 90),    # Hot
    (90, 100),   # Very Hot
    (100, 150)   # Extremely Hot
]

def categorize_temperature(temp):
    for i, (lower, upper) in enumerate(temperature_ranges):
        if lower <= temp < upper:
            return i
    return len(temperature_ranges) - 1

df['Temperature_Range'] = df['Temperature'].apply(categorize_temperature)

grouped_data = df.groupby(['Day_of_Week', 'Temperature_Range']).size().unstack(fill_value=0)

plt.figure(figsize=(10, 6))
plt.imshow(grouped_data, cmap='Blues', interpolation='nearest', aspect='auto')

plt.colorbar(label='Number of Accidents')

plt.xlabel('Temperature Range')
plt.ylabel('Day of the Week')
plt.xticks(np.arange(len(temperature_ranges)), [f'{temp[0]}-{temp[1]}' for temp in temperature_ranges], rotation=45)
plt.yticks(np.arange(7), ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])

plt.title('Number of Accidents by Day of the Week and Temperature Range')

plt.tight_layout()
plt.show()
````

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Convert 'Start_Time' column to datetime format
df['Start_Time'] = pd.to_datetime(df['Start_Time'])

# Create a new column 'Day_Night' to categorize accidents as 'Day' or 'Night'
df['Day_Night'] = pd.cut(df['Start_Time'].dt.hour, bins=[0, 6, 18, 23], labels=['Night', 'Day', 'Night_End'])
# Updated labels to 'Night_End' for hours 18-23

# Create a boxplot to compare temperature for day and night accidents
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Temperature', y='Day_Night', hue='Day_Night', palette='Set2', orient='h', dodge=False)
plt.title('Temperature Patterns for Day and Night Accidents')
plt.xlabel('Temperature (F)')
plt.ylabel('Day/Night')
plt.show()
````

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt



# Convert 'Start_Time' column to datetime format
df['Start_Time'] = pd.to_datetime(df['Start_Time'])

# Extract hour of the day
df['Hour'] = df['Start_Time'].dt.hour

# Create a line plot to show temperature fluctuations during sunrise and sunset
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Hour', y='Temperature', hue='Sunrise_Sunset', errorbar=None)
plt.title('Temperature Fluctuations during Sunrise and Sunset')
plt.xlabel('Hour of the Day')
plt.ylabel('Temperature (F)')
plt.xticks(range(0, 24))
plt.legend(title='Sunrise_Sunset')
plt.show()
````

### Weather Condition

````python
import pandas as pd



# Select the columns for correlation analysis
weather_columns = ['Temperature', 'Wind_Chill(F)', 'Humidity(%)', 'Pressure(in)', 
                   'Visibility(mi)', 'Wind_Speed(mph)', 'Precipitation(in)']

# Calculate correlation coefficients with 'Severity'
weather_corr = df[weather_columns + ['Severity']].corr()

# Display correlation matrix
print("Correlation Matrix between Weather Variables and Accident Severity:")
print(weather_corr)
````

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(weather_corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Correlation Matrix between Weather Variables and Accident Severity")
plt.show()
````

````python
import seaborn as sns
import matplotlib.pyplot as plt


# Create a countplot of accidents by weather condition
plt.figure(figsize=(12, 8))
sns.countplot(data=df, x='Weather_Condition', hue='Weather_Condition', 
              order=df['Weather_Condition'].value_counts().index[:20], 
              palette='viridis', legend=False)
plt.title('Number of Accidents by Weather Condition')
plt.xlabel('Weather Condition')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
````

````python
# 1 Are there more accidents in warmer areas than colder areas?

cold_cities=['Fairbanks','Grand Forks','Williston','Fargo','Aberdeen']
warm_cities=[ 'Mesa', 'Arizona', 'Honolulu', 'Hawaii', 'Orlando']
city_counts = df['City'].value_counts()

cold = 0
for city in cold_cities:
    cold += city_counts.get(city, 0)

warm=0
for city in warm_cities:
    warm += city_counts.get(city,0)


print(f"Total appearances of cold cities: {cold} & Avg : {cold/5}")
print(f"Total appearances of warm cities:,{warm} & Avg : {warm/5}")


````

- ``` According to the top 5 hottest and coldest cities accident average we Can say that in Warmer cities most of the accidents happens ```

````python
#2 Which 5 states have the highest number of accidents? How about per capita?

Top5_state=df.State.value_counts()
print("Top 5 states with highest number of accidents:\n",Top5_state[:5])

per_capita=['DC','MA','CT','NJ','CA']
state_counts = df['State'].value_counts()

per_capita_sorted = sorted(per_capita, key=lambda x: state_counts.get(x, 0), reverse=True)

print("Top 5 States with Highest number of accidents from per capita POV:\n",per_capita_sorted)
````
